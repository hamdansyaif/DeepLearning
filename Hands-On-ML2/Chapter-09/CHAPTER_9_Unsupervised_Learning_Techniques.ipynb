```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

**Using Clustering for Image Segmentation**

Segmentasi gambar adalah tugas untuk membagi gambar menjadi beberapa segmen. Dalam segmentasi semantik, semua piksel yang merupakan bagian dari jenis objek yang sama akan ditetapkan ke segmen yang sama. Misalnya, dalam sistem penglihatan mobil tanpa pengemudi, semua piksel yang merupakan bagian dari gambar pejalan kaki mungkin ditetapkan ke segmen "pejalan kaki" (akan ada satu segmen yang berisi semua pejalan kaki). Dalam segmentasi instan, semua piksel yang merupakan bagian dari objek individual yang sama ditetapkan ke segmen yang sama. Dalam kasus ini akan ada segmen yang berbeda untuk setiap pejalan kaki. Keadaan terkini dalam segmentasi semantik atau instan saat ini dicapai dengan menggunakan arsitektur kompleks yang didasarkan pada jaringan saraf konvolusional (lihat Bab 14). Di sini, kita akan melakukan sesuatu yang jauh lebih sederhana: segmentasi warna. Kita hanya akan menetapkan piksel ke segmen yang sama jika mereka memiliki warna yang sama. Dalam beberapa aplikasi, ini mungkin cukup. Misalnya, jika Anda ingin menganalisis citra satelit untuk mengukur berapa banyak total area hutan di suatu wilayah, segmentasi warna mungkin baik-baik saja. Beberapa gambar mungkin memiliki lebih sedikit saluran, seperti gambar skala abu-abu (satu saluran). Dan beberapa gambar mungkin memiliki lebih banyak saluran, seperti gambar dengan saluran alfa tambahan untuk transparansi atau gambar satelit, yang sering kali berisi saluran untuk banyak frekuensi cahaya (misalnya, inframerah). Kode berikut membentuk ulang array untuk mendapatkan daftar panjang warna RGB, lalu mengelompokkan warna-warna ini menggunakan K-Means:

```python
X = image.reshape(-1, 3)
n_colors = (10, 8, 6, 4, 2)
kmeans = KMeans(n_clusters=8).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 3, 1)
plt.title("Original image")
plt.imshow(image)
plt.axis('off')

for i, k in enumerate(n_colors):
    print(f"Memproses untuk k={k} warna...")

    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X)

    new_colors = kmeans.cluster_centers_
    segmented_img_flat = new_colors[kmeans.labels_]
    segmented_img = segmented_img_flat.reshape(image.shape)

    # Perulangan ini akan mengisi posisi 2, 3, 4, 5, dan 6
    plt.subplot(2, 3, i + 2)
    plt.title(f"{k} colors")
    plt.imshow(segmented_img)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

Ini menghasilkan gambar yang ditunjukkan di kanan atas Gambar 9-12. Anda dapat bereksperimen dengan berbagai jumlah klaster, seperti yang ditunjukkan pada gambar. **Using Clustering for Preprocessing**

Pengelompokan dapat menjadi pendekatan yang efisien untuk pengurangan dimensionalitas, khususnya sebagai langkah praproses sebelum algoritma pembelajaran terbimbing. Sebagai contoh penggunaan pengelompokan untuk pengurangan dimensionalitas, mari kita bahas kumpulan data digit, yang merupakan kumpulan data sederhana seperti MNIST yang berisi 1.797 gambar skala abu-abu 8  8 yang merepresentasikan digit 0 hingga 9. Pertama, muat kumpulan data:

```python
 from sklearn.datasets import load_digits
 X_digits, y_digits = load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([("kmeans", KMeans(n_clusters=50)),("log_reg", LogisticRegression()),])
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = dict(kmeans__n_clusters=range(2, 100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test, y_test)
```

**Using Clustering for Semi-Supervised Learning**

Kasus penggunaan lain untuk pengelompokan adalah dalam pembelajaran semi-supervised, ketika kita memiliki banyak
instance yang tidak berlabel dan sangat sedikit instance yang berlabel. Mari kita latih model Regresi Logistik
pada sampel 50 instance berlabel dari kumpulan data digit

```python
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
log_reg.score(X_test, y_test)
```

```python
k = 50
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
X_digits_dist = kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train[representative_digit_idx]
```

image.png

Gambar 9-13. Lima puluh gambar digit representatif (satu per klaster)

```python
y_representative_digits = y_train[representative_digit_idx]
```

```python
print(f"Jumlah gambar representatif: {X_representative_digits.shape[0]}")
print(f"Jumlah label representatif: {y_representative_digits.shape[0]}")
```

```python
log_reg = LogisticRegression(max_iter=2000, random_state=42)
log_reg.fit(X_representative_digits, y_representative_digits)
```

```python
score = log_reg.score(X_test, y_test)
print(f"\nSkor akurasi pada test set: {score:.4f}")
```

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
 y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

```python
percentile_closest = 20

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
 in_cluster = (kmeans.labels_ == i)
 cluster_dist = X_cluster_dist[in_cluster]
 cutoff_distance = np.percentile(cluster_dist, percentile_closest)
 above_cutoff = (X_cluster_dist > cutoff_distance)
 X_cluster_dist[in_cluster & above_cutoff] = -1
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
np.mean(y_train_partially_propagated == y_train[partially_propagated])
```

**DBSCAN**

Algoritma ini mendefinisikan kluster sebagai wilayah berkesinambungan dengan kepadatan tinggi. Berikut cara kerjanya:


*  Untuk setiap contoh, algoritma menghitung berapa banyak contoh yang terletak dalam jarak kecil  (epsilon) darinya. Wilayah ini disebut neighborhood contoh tersebut.
*   Jika sebuah instance memiliki setidaknya instance min_samples di lingkungan -nya (termasuk dirinya sendiri), maka instance tersebut dianggap sebagai instance inti. Dengan kata lain, instance inti adalah instance yang terletak di wilayah padat.
- Semua instans di sekitar instans inti termasuk dalam kluster yang sama.
Lingkungan ini dapat mencakup instans inti lainnya; oleh karena itu, rangkaian panjang instans inti yang berdekatan membentuk kluster tunggal.



```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_
```

Pengelompokan ini direpresentasikan dalam plot sebelah kiri Gambar 9-14. Seperti yang dapat Anda lihat, diagram ini mengidentifikasi banyak anomali, ditambah tujuh klaster yang berbeda. Sungguh mengecewakan! Untungnya, jika kita memperluas lingkungan setiap contoh dengan meningkatkan eps menjadi 0,2, kita mendapatkan pengelompokan di sebelah kanan, yang tampak sempurna. Mari kita lanjutkan dengan model ini.

image.png

Gambar 9-14. Pengelompokan DBSCAN menggunakan dua radius lingkungan yang berbeda

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
```

```python
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2] = -1
y_pred.ravel()
```

**Anomaly Detection Using Gaussian Mixtures**

Deteksi anomali (juga disebut deteksi outlier) adalah tugas mendeteksi kejadian yang sangat menyimpang dari norma. Kejadian ini disebut anomali, atau outlier, sedangkan kejadian normal disebut inlier. Deteksi anomali berguna dalam berbagai aplikasi, seperti deteksi penipuan, mendeteksi produk cacat dalam produksi, atau menghilangkan outlier dari kumpulan data sebelum melatih model lain (yang dapat meningkatkan kinerja model yang dihasilkan secara signifikan). Menggunakan model campuran Gaussian untuk deteksi anomali cukup sederhana: setiap kejadian yang terletak di wilayah dengan kepadatan rendah dapat dianggap sebagai anomali. Anda harus menentukan ambang batas kepadatan yang ingin Anda gunakan. Misalnya, di perusahaan manufaktur yang mencoba mendeteksi produk cacat, rasio produk cacat biasanya sudah diketahui. Katakanlah sama dengan 4%. Anda kemudian menetapkan ambang batas kepadatan menjadi nilai yang menghasilkan 4% kejadian yang terletak di area di bawah kepadatan ambang batas tersebut. Jika Anda melihat bahwa Anda mendapatkan terlalu banyak hasil positif palsu (misalnya, produk yang sangat bagus tetapi ditandai sebagai cacat), Anda dapat menurunkan ambang batas. Sebaliknya, jika Anda memiliki terlalu banyak hasil negatif palsu (misalnya, produk cacat yang tidak ditandai sebagai cacat oleh sistem), Anda dapat meningkatkan ambang batas. Ini adalah trade-off presisi/pengingatan yang biasa (lihat Bab 3). Berikut adalah cara mengidentifikasi outlier menggunakan kepadatan terendah persentil keempat sebagai ambang batas (misalnya, sekitar 4% dari contoh akan ditandai sebagai anomali):

```python
densities = gm.score_samples(X)
density_threshold = np.percentile(densities, 4)
anomalies = X[densities < density_threshold]
```

Misalnya, mari kita tetapkan jumlah klaster menjadi 10 dan lihat apa yang terjadi:

```python
from sklearn.mixture import BayesianGaussianMixture

bgm = BayesianGaussianMixture(n_components=10, n_init=10)
bgm.fit(X)
np.round(bgm.weights_, 2)
```
